```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
# read the csv file
recent_grads = pd.read_csv('recent-grads.csv')
recent_grads = recent_grads.dropna()
# check women and men for different major categories
df = recent_grads.groupby(['Major_category'])[["Women", "Men"]].sum()
# draw a plot to compare women and men for major categories
df1 = df['Women']
df2 = df['Men']
plt.figure(figsize=(10,5))
ax = df1.plot.bar(color = 'r', label='women')
df2.plot.bar(color = 'b', bottom = df1, ax=ax, label='men')
plt.title('Women and Men by category')
plt.ylabel('Amount')
plt.legend(loc='upper right')
plt.show()
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
# read the csv file
recent_grads = pd.read_csv('recent-grads.csv')
recent_grads = recent_grads.dropna()
df = recent_grads.groupby(['Major_category'])[["Unemployment_rate"]].mean()
df2 = df['Unemployment_rate']
plt.figure(figsize=(10,5))
ax = df2.plot.bar(color = 'red', label='unemployment')
plt.title('Unemployment by major category')
plt.ylabel('Unemployment rate')
plt.show()

```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
# read the csv file
recent_grads = pd.read_csv('recent-grads.csv')
recent_grads = recent_grads.dropna()
df = recent_grads.groupby(['Major_category'])[["Median"]].mean()
df2 = df['Median']
plt.figure(figsize=(10,5))
ax = df2.plot.bar(color = 'green', label='unemployment')
plt.title('Median salary by major category')
plt.ylabel('Salary $')
plt.show()
```
